```python
from datetime import datetime, timezone
import numpy as np
import av
import av.datasets
import glob
import os
import pandas as pd
from tqdm import tqdm
```

```python
folder = r'C:\Users\dmchacon\Documents\OAK-D Videos\3_7\2-14442C107152C0D200'
videos = glob.glob(os.path.join(folder,'*.mp4'))
videos
```

```python
video = videos[0]
content = av.datasets.curated(video)

out_folder = video.split('.mp4')[0]
os.makedirs(out_folder,exist_ok=True)

with av.open(content) as container:
    # Signal that we only want to look at keyframes.
    stream = container.streams.video[0]
    stream.codec_context.skip_frame = "NONKEY"
    for frame in container.decode(stream):
        # We use `frame.pts` as `frame.index` won't make must sense with the `skip_frame`.
        frame.to_image().save(
            os.path.join(out_folder,"{:04d}.png".format(frame.pts)),
            quality=80,
        )
```
