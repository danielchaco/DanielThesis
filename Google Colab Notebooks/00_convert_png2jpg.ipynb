```python
folder = '/content/drive/MyDrive/Daniel_Thesis/01_Data/Scans'
data = []
for root, dirs, files in os.walk(folder, topdown=False):
   for name in files:
      data.append([os.path.join(root, name),name.split('.')[-1]])
df = pd.DataFrame(data,columns=['path','ftype'])
```

```python
df[df.ftype=='png']
```

```python
out_folder = '/content/drive/MyDrive/Daniel_Thesis/01_Data/Scans_jpg'
os.makedirs(out_folder,exist_ok=True)
for path in tqdm(df[df.ftype=='png'].path):
  image = cv2.imread(path)
  foldername = path.split('/')[-2]
  name = os.path.basename(path).split('.')[0]
  cv2.imwrite(os.path.join(out_folder,foldername.strip()+'_'+name+'.jpg'), image, [int(cv2.IMWRITE_JPEG_QUALITY), 100])
```
